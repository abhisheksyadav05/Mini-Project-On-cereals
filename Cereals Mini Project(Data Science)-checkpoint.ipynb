```python
import pandas as pd
import numpy as np
import seaborn as sns
import random as rd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

```

```python
data = pd.read_csv('Cereals.csv')
data.head(40)
```

```python
 datax= data.drop(['Name','Rating'], axis = 1)
```

```python
#Shape of data
datax.head(20)
```

```python
datax.shape
```

```python
#getting some info
datax.info()
```

```python
# checking for missing values
datax.isnull().sum()
```

```python
datay = datax.fillna(0)
```

```python
datay.isnull().sum()
```

```python
#chossing the calories and cups coloumn
X = datay.iloc[:,[2,13]].values
print(X)
```

```python
#wcss 
#elbow method
wcss = []

for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++', random_state = 20)
    kmeans.fit(X)
    
    wcss.append(kmeans.inertia_)
    
    
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow point graph')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#optimum no of cluster =5
#training the k-means clustering model
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)

# return a label for each data point based on there cluster
Y= kmeans.fit_predict(X)
print(Y)

```

```python
#visualising all the clusters
#plotting all the clusters and there centroids
plt.figure(figsize=(8,8))
plt.scatter(X[Y== 0,0], X[Y==0,1], s=50, c= 'green', label= 'cluster 1')
plt.scatter(X[Y== 1,0], X[Y==1,1], s=50, c= 'red', label= 'cluster 2')
plt.scatter(X[Y== 2,0], X[Y==2,1], s=50, c= 'yellow', label= 'cluster 3')
plt.scatter(X[Y== 3,0], X[Y==3,1], s=50, c= 'orange', label= 'cluster 4')
plt.scatter(X[Y== 4,0], X[Y==4,1], s=50, c= 'blue', label= 'cluster 5')

#plot the centroid
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', label = 'Centroids')
plt.title('Clusters of calories')
plt.xlabel('calories')
plt.ylabel('cups')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)

# return a label for each data point based on there cluster
Z= kmeans.fit_predict(X)
print(Z)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[Z== 0,0], X[Z==0,1], s=50, c= 'green', label= 'cluster 1')
plt.scatter(X[Z== 1,0], X[Z==1,1], s=50, c= 'red', label= 'cluster 2')
plt.scatter(X[Z== 2,0], X[Z==2,1], s=50, c= 'yellow', label= 'cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', label = 'Centroids')
plt.title('Clusters of calories')
plt.xlabel('calories')
plt.ylabel('cups')
plt.show()
```
